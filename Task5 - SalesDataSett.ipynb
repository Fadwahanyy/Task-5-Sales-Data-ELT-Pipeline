# Sales by Territory
This chart shows how sales/orders are distributed across different territories (EMEA, APAC, Japan, etc.).


```python
x = df['TERRITORY'].value_counts(sort=False).index
y = df['TERRITORY'].value_counts(sort=False).values

plt.bar(x, y)

plt.xlabel('TERRITORY')
plt.ylabel('Count')

plt.show()
```

# Sales Distribution
We analyze the overall distribution of sales values to detect skewness and potential outliers.



```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')

monthly_sales = df.groupby(pd.Grouper(key='ORDERDATE', freq='M'))['SALES'].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(7,4))
sb.boxplot(x=df['SALES'])
plt.title("Sales Distribution (Boxplot)")
plt.xlabel("Sales")
plt.tight_layout()
plt.show()
```

# Sales by Product Line
We compare total sales across different product categories to identify the top performers.


```python
product_sales = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
product_sales.plot(kind='bar', color='lightgreen')
plt.title("Total Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Sales by Deal Size
This section examines orders grouped by deal size (Small, Medium, Large) to understand order volume distribution.


```python
sorted_counts = df['DEALSIZE'].value_counts()
gen_percentage = sorted_counts/sorted_counts[:].sum()
plt.pie(gen_percentage, labels = sorted_counts.index, startangle = 90, counterclock = False);
```
